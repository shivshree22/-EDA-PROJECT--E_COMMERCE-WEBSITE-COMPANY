#Que5.	Produce a list of all the unique values & data types present in the non-numeric columns in df.?

```python
#select non numeric data type column from df 
non_num_cl=df.describe(exclude="number")

# Create a dictionary to store unique values and data types

unique_values_dtypes={}

# Iterate over each non-numeric column by non_num_cl using for loop and get seprate as per unique values

for column in non_num_cl:
    
    # get unique values from dataset
    unique_values= df[column].unique()
    # get column dtypes from non_num_cl
    col_dtypes=df[column].dtypes
 
    # now store that all values and types in dict 

    unique_values_dtypes[column]={"unique_values":unique_values,"col_dtypes":col_dtypes}
    
# Print the unique values and data types
for column, values_types in unique_values_dtypes.items():
    print(f"Column: {column}")
    print(f"Data Type: {values_types['col_dtypes']}")
    print(f"Unique Values: {values_types['unique_values']}")
    print()    
```

#Que1.	What are the top 3 “Place_in_India”  on the basis of column “Level 4” for the year 2020 and 2022 separately ?

```python
# first filter or lock the data for 2020

a=df.loc[df["Year"]==2020]


#now time groupby as per place in india with level 4
group_2020=a.groupby(["Year","Place_in_India"]).agg({"Level 4":"sum"})

# sort the city as per their high sum values

top3_city_2020=group_2020.sort_values(by="Level 4",ascending= False).head(3)


#find rank of each city by top order 
# BY USING RANGE FUNCTION AND use of len of rows in data  we create a range from 1 to end and assign values to rank 

top3_city_2020["Rank"]=range(1,len(top3_city_2020)+1)

  
# for that reset index and add column to data 

top3_city_2020=top3_city_2020.reset_index()


# now select only desired columns using index opreator  and assign back to dataframe 

top3_city_2020=top3_city_2020[["Year","Rank","Place_in_India","Level 4"]]

# Rename the columns
top3_city_2020 = top3_city_2020.rename(columns={ "Level 4": "Sum of Level 4"})

#######now do all same procedure for year 2022

# first filter or lock the data for 2020

a1=df.loc[df["Year"]==2022]


#now time groupby as per place in india with level 4
group_2022=a1.groupby(["Year","Place_in_India"]).agg({"Level 4":"sum"})

# sort the city as per their high sum values

top3_city_2022=group_2022.sort_values(by="Level 4",ascending= False).head(3)


#find rank of each city by top order 
# BY USING RANGE FUNCTION AND use of len of rows in data  we create a range from 1 to end and assign values to rank 

top3_city_2022["Rank"]=range(1,len(top3_city_2022)+1)

# now one column is remaning that city  
# for that reset index and add column to data 

top3_city_2022=top3_city_2022.reset_index()


# now select only desired columns using index opreator  and assign back to dataframe 

top3_city_2022=top3_city_2022[["Year","Rank","Place_in_India","Level 4"]]

# Rename the columns
top3_city_2022 = top3_city_2022.rename(columns={ "Level 4": "Sum of Level 4"})


# Print the top cities in 2020
print("Top Cities in India for the year 2020:")
print(top3_city_2020.head(3))
print()
print()
# Print the top cities in 2020
print("Top Cities in India for the year 2022:")
print(top3_city_2022.head(3))
```

#Que3.	What are the bottom 3 “Place_in_India”  on the basis of column “Level 4”/ “Level 1” for the year 2021 and 2022 separately ?

```python
# first loc data by year 

df_2021 = df.loc[df["Year"]==2021]

#create a column  in data “Level 4”/ “Level 1”

df_2021["Level 4/ Level 1"]=df["Level 4"]/df["Level 1"]

#group by place of india 

df_2021=df_2021.groupby(["Place_in_India"]).agg({"Level 4/ Level 1":"sum"})

# we want last 3 as per level in palce in india  for that e use sort as per


bottom3_df_2021=df_2021.sort_values(by="Level 4/ Level 1",ascending=False)

# reset index

bottom3_df_2021=bottom3_df_2021.reset_index()

# assign column 

bottom3_df_2021=bottom3_df_2021[["Place_in_India","Level 4/ Level 1"]]



##### done same as for year 2022


# first loc data by year 

df_2022 = df.loc[df["Year"]==2022]

#create a column  in data “Level 4”/ “Level 1”

df_2022["Level 4/ Level 1"]=df["Level 4"]/df["Level 1"]

#group by place of india 

df_2022=df_2022.groupby(["Place_in_India"]).agg({"Level 4/ Level 1":"sum"})

# we want last 3 as per level in palce in india  for that e use sort as per


bottom3_df_2022=df_2022.sort_values(by="Level 4/ Level 1",ascending=False)

# reset index

bottom3_df_2022=bottom3_df_2022.reset_index()

# assign column 

bottom3_df_2022=bottom3_df_2022[["Place_in_India","Level 4/ Level 1"]]

# Print the bottom cities in 2022
print("Top Cities in India for the year 2021:")
print(bottom3_df_2021.tail(3))
print()
print()
# Print the bottom cities in 2022
print("Top Cities in India for the year 2022:")
print(bottom3_df_2022.tail(3))
```
